## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
##### Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("title")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.


####  Use the following data:
#### import numpy as np
#### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
#### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#### The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots in a grid layout, making it easier to compare and analyze different visualizations.

#### The subplot() function takes three arguments: nrows, ncols, and index. Here's the syntax:

```python
import matplotlib.pyplot as plt
#For line 1:
plt.subplot(2, 2, 1)
plt.plot([0, 1, 2, 3, 4, 5], [0, 100, 200, 300, 400, 500])
plt.title('Subplot 1')

#For line 2:
plt.subplot(2, 2, 2)
plt.plot([0, 1, 2, 3, 4, 5], [50, 20, 40, 20, 60, 70])
plt.title('Subplot 2')

#For line 3:
plt.subplot(2, 2, 3)
plt.plot([0, 1, 2, 3, 4, 5], [10, 20, 30, 40, 50, 60])
plt.title('Subplot 3')

#For line 4:
plt.subplot(2, 2, 4)
plt.plot([0, 1, 2, 3, 4, 5],[200, 350, 250, 550, 450, 150])
plt.title('Subplot 4')

plt.tight_layout()
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.



```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

### A bar plot, also known as a bar chart, is a type of chart that uses rectangular bars to represent categorical data. It displays the relationship between a categorical variable and a numerical value or frequency.

#### Bar plots are commonly used to: Compare Data, Display Distribution, Show Patterns or Trends,Present Results.

```python
plt.bar(company, profit)
plt.show()
```

#### Box plots are used for purposes like: Visualizing Distribution, Comparing Groups, Detecting Outliers,Assessing Skewness.
